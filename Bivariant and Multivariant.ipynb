```python
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic.csv')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
titanic.head(5)

tips.head(5)

flights.head(5)

iris.head(5)
```

# Numerical - Numerical 

### scatterplot

```python
sns.scatterplot(data = tips, x = tips['total_bill'], y = tips['tip'])
```

```python
sns.scatterplot(data = tips, x = tips['total_bill'], y = tips['tip'], hue = tips['sex'])
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker')
```

```python
sns.scatterplot(data = tips, x = 'total_bill', y = 'tip', hue = 'sex', style = 'smoker', size = 'size')
```

### pairplot 

```python
sns.pairplot(data = iris, hue = 'species')
```

```python
sns.pairplot(data = tips, hue = 'sex')
```

### lineplot    (used when we have time series (plot time on x-axis))

```python
sns.lineplot(data = flights, x = 'year', y = 'passengers')
```

```python
new = flights.groupby('year').sum(numeric_only=True)

sns.lineplot(data = new, x = 'year', y = 'passengers')
```

# Categorial - Categorial

### Heatmap

```python
sns.heatmap(pd.crosstab(tips['sex'], tips['smoker']))
```

```python
sns.heatmap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
sns.heatmap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

### clustermap

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
sns.clustermap(flights.pivot_table(values = 'passengers', index = 'month', columns = 'year'))
```

# Categorical - Numerical

### barplot

```python
# titanic.groupby('Pclass')['Age'].mean().plot(kind = 'bar')

sns.barplot(data = titanic, x = 'Pclass', y = 'Age', hue = 'Sex')
```

```python
sns.barplot(data = titanic, x = 'Pclass', y = 'Survived')
```

### displot & histplot

```python
survived = titanic[titanic['Survived'] == 1]
not_survived = titanic[titanic['Survived'] == 0]

sns.histplot(data = survived, x = 'Age', kde = True)
sns.histplot(data = not_survived, x = 'Age', kde = True)
```

### kdeplot             (only for KDE line)

```python
survived = titanic[titanic['Survived'] == 1]
not_survived = titanic[titanic['Survived'] == 0]

sns.kdeplot(data = survived, x = 'Age')
sns.kdeplot(data = not_survived, x = 'Age')
```

### boxplot

```python
sns.boxplot(data = titanic, x = 'Sex', y = 'Age')
```
